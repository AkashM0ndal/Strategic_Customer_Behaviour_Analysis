```python
# import essential library for python
import pandas as pd
```

```python
# retrieve the dataset into python for analysis
df = pd.read_csv(r"C:\Users\am647\Desktop\WOrking on Datasets\customer_shopping_behavior.csv")
```

```python
# checking the how many missing values have in each column

df.isnull().sum()
```

```python
# fill the NaN values in 'review rating' column with the median value of each category over the mean value of the review ratings
# if there present outliers in the mean the our dataset becomes biased
# and also we put median value by each category
# this method is more professional to 



df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# checking the NaN values are present or not in 'Review Rating' Column

df.isnull().sum()
```

```python
# turn all the column names into snake casing,all the letters in small letter and instead of spaces put '_'

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
# checking the upper code runs currect or not

df.columns
```

```python
# rename column for easy to write the name of the column in the code

df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
# checking the column name change or not 

df.columns
```

```python
# create a column 'age_group'

age_group_labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4 , labels = age_group_labels)
```

```python
df.columns
```

```python
# retrieve both column to see the result

df[['age','age_group']].head(10)
```

```python
# See the frequency_of_purchases column vlaues to turn into digits for easy to analysis
df['frequency_of_purchases'].head(10)
```

```python
# see the unique values in frequency_of_purchases

set(df['frequency_of_purchases'])
```

```python
# put some integer values instead of frequency_of_purchases column values

frequency_mapping = {
            'Annually' : 365,
            'Bi-Weekly' : 14,
            'Every 3 Months' : 90,
            'Fortnightly' : 14,
            'Monthly' : 30,
            'Quarterly' : 90,
            'Weekly' : 7
        
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# see the upper code runs perfectly or not

df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# check discount_applied and promo_code_used both column values are identically same or not

df[['discount_applied','promo_code_used']].head(10)
```

```python
# check discount_applied and promo_code_used both column values are identically same or not

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# delete the promo_code_used column to filter out the same column which have twice

df.drop('promo_code_used',axis = 1, inplace = True)
```
